```python
# Import dependencies
import pandas as pd

# Read CSV file
income = pd.read_csv('Resources/Raw Data/zipPopDensity_meanHHI.csv')
```

```python
# Convert to DataFrame
income_df = pd.DataFrame(income)
income_df
```

```python
# Drop unneccesary columns
income_df.drop(income_df.columns[[1, 2]], axis=1, inplace=True)

# Rename columns
income_df = income_df.rename(columns = {'ZIP_CODE':'Zip Code', 'zipCodePop':'Zip Population', 
                                        'areaSqMile':'Zip SqMi', 'PopDensity':'Zip Pop Density',
                                        'MeaHHI':'Zip Mean HHI'})

income_df
```

```python
income_df.isna().sum()
```

```python
# Drop rows with empty cells 
income_df = income_df.dropna()
```

```python
income_df.isna().sum()
```

```python
# Export to csv
income_df.to_csv('Resources/Clean Data/clean_income.csv', index=False)
```

```python
# Import DataFrame into database
from sqlalchemy import create_engine
from config import db_password
db_string = f"postgresql://postgres:{db_password}@127.0.0.1:5432/Portland_Housing"
engine = create_engine(db_string)
income_df.to_sql(name='income', con=engine)
```
